```python
data=pd.read_csv(r"C:/Users/hp/Downloads/youtube_subscribers_data.csv")
```

```python
data.info()
```

```python
data['Subscribers (millions)'] = pd.to_numeric(data['Subscribers (millions)'],errors='coerce')
```

```python
data
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['Subscribers (millions)'])
```

```python
#Removng last row which contains unnecessary data
data.drop(data[data['Name'] == 'As of December 2024 (UTC)'].index,inplace=True)
```

```python
data.drop(data.columns[[1]],axis=1,inplace=True)
```

```python
data.drop(data.columns[[-1]],axis=1,inplace=True)
```

```python
data['Primary language']=data['Primary language'].map(lambda x : x.split('[')[0])
data['Category'] = data['Category'].str.split('[').str[0]
```

```python
data.to_csv('youtube_subscribers_data.csv',index=False)
```
